### Univariate Analysis

```python
# Univariate Analysis: Analyze single variable distributions
def univariate_analysis(df):
    # Numerical features distribution
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    print(f"\nUnivariate analysis on numerical columns: {numerical_cols}")
    
    # Plotting histograms for numerical features
    df[numerical_cols].hist(bins=20, figsize=(30, 15))
    plt.suptitle("Distribution of Numerical Features")
    plt.show()

    # Categorical features distribution
    categorical_cols = df.select_dtypes(include=[np.object_]).columns
    print(f"\nUnivariate analysis on categorical columns: {categorical_cols}")

    # Plotting bar plots for categorical features
    for col in categorical_cols:
        plt.figure(figsize=(8, 5))
        sns.countplot(data=df, x=col, palette='Set2')
        plt.title(f"Distribution of {col}")
        plt.xticks(rotation=45)
        plt.show()

univariate_analysis(data_clean)
```

### Bivariate Analysis

```python
# Bivariate Analysis: Analyze relationships between two variables
def bivariate_analysis(df):
    print("\nBivariate analysis between Status and other variables:")
    
    # Categorical vs Status
    categorical_cols = df.select_dtypes(include=[np.object_]).columns
    for col in categorical_cols:
        if col != 'Status':
            plt.figure(figsize=(8, 5))
            sns.countplot(data=df, x=col, hue='Status', palette='Set1')
            plt.title(f"{col} vs Status")
            plt.xticks(rotation=45)
            plt.show()

    # Numerical vs Status
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    for col in numerical_cols:
        if col != 'Status':
            plt.figure(figsize=(8, 5))
            sns.boxplot(data=df, x='Status', y=col, palette='Set2')
            plt.title(f"{col} vs Status")
            plt.show()

bivariate_analysis(data_clean)
```

### Multivariate Analysis

```python
# Multivariate Analysis: Analyze relationships across multiple variables
def multivariate_analysis(df):
    print("\nMultivariate analysis (correlation heatmap):")
    
    # Correlation matrix for numerical features
    numeric_columns = ['Age_at_enrollment', 'Curricular_units_1st_sem_credited', 
                       'Curricular_units_1st_sem_approved', 'Curricular_units_1st_sem_evaluations', 
                       'Admission_grade', 'Debtor', 'Tuition_fees_up_to_date']
    
    # Calculate correlation matrix for the selected numerical features
    correlation_matrix = df[numeric_columns].corr()
    
    # Create heatmap to visualize the correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title("Correlation Heatmap of Numerical Features")
    plt.show()

# Select numerical columns from the dataframe for analysis
numeric = df.select_dtypes(include=[float, int])

# Run multivariate analysis on the selected numerical features
multivariate_analysis(numeric)
```

### Check Outliers

```python
# Checking for Outliers
def check_outliers(df):
    print("\nChecking for Outliers using IQR:")
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    for col in numerical_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1

        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        if len(outliers) != 0:
            print(f"Outliers in {col}: {len(outliers)} out of {df.shape[0]} rows")

check_outliers(data_clean)
```

### Split Dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
X = df.drop(
['Status', 
'Daytime_evening_attendance',
'Curricular_units_2nd_sem_evaluations',
'Fathers_occupation', 'Mothers_occupation',
'Fathers_qualification', 'Mothers_qualification',
'Curricular_units_1st_sem_evaluations',
'Curricular_units_1st_sem_without_evaluations',
'Curricular_units_2nd_sem_without_evaluations',
'Application_mode',
'Application_order', 'Course',
'Previous_qualification', 'Admission_grade',
'Previous_qualification_grade'], 
axis=1)

Y = df['Status']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


# Buat instance MinMaxScaler
scaler = MinMaxScaler()

# Fit dan transform data training
X_train_scaled = scaler.fit_transform(X_train)

# Transform data testing
X_test_scaled = scaler.transform(X_test)
```

### XGBoost

```python
import xgboost as xgb
from sklearn import metrics
```

```python
# Train an XGBoost model
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, Y_train)
```

### Random Forest

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train a Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)
```

## Evaluation

### XGBoost

```python
from sklearn import metrics

# Make predictions on the test set
Y_pred = xgb_model.predict(X_test)

# Show the Accuracy, Precision, Recall, F1 Score, and Cohen's Kappa
accuracy_gnb = metrics.accuracy_score(Y_test, Y_pred)
precision_gnb = metrics.precision_score(Y_test, Y_pred, average='macro')  
recall_gnb = metrics.recall_score(Y_test, Y_pred, average='macro')
f1_gnb = metrics.f1_score(Y_test, Y_pred, average='macro')
kappa_gnb = metrics.cohen_kappa_score(Y_test, Y_pred)

print("Accuracy:", accuracy_gnb)
print("Precision:", precision_gnb)
print("Recall:", recall_gnb)
print("F1 Score:", f1_gnb)
print("Cohens Kappa Score:", kappa_gnb)

# Confusion Matrix
print(metrics.confusion_matrix(Y_test, Y_pred))
```

### Random Forest

```python
# Make predictions on the test set
Y_pred = rf_model.predict(X_test)

# Show the Accuracy, Precision, Recall
accuracy_gnb = metrics.accuracy_score(Y_test, Y_pred)
precision_gnb = metrics.precision_score(Y_test, Y_pred, average='macro')
recall_gnb = metrics.recall_score(Y_test, Y_pred, average='macro')
f1_gnb = metrics.f1_score(Y_test, Y_pred, average='macro')
kappa_gnb = metrics.cohen_kappa_score(Y_test, Y_pred)

print("Accuracy:", accuracy_gnb)
print("Precision:", precision_gnb)
print("Recall:", recall_gnb)
print("F1 Score:", f1_gnb)
print("Cohens Kappa Score:", kappa_gnb)

print(metrics.confusion_matrix(Y_test, Y_pred))
```
